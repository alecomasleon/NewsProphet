```python
import sys
sys.path.append('/usr/local/lib/python3.9/site-packages/')
sys.path.append('/backend/model_creation')

from ucimlrepo import fetch_ucirepo
from data_scraping import scrape_article
from get_features import get_features_from_article_data
from mashable_url import make_valid_mashable_url
import pandas as pd
import get_features
```

```python
online_news_popularity = fetch_ucirepo(id=332) 
```

```python
data = online_news_popularity.data.original
urls = data['url'].to_list()
y = online_news_popularity.data.targets[' shares'].to_list()
```

```python
entry = get_features_from_article_data(scrape_article(make_valid_mashable_url(urls[0])), do_flair=False)
entry.insert(0, "url", [urls[0]], True)
entry.insert(len(entry.columns), 'shares', y[0])
entry.to_csv('raw_extracted_data.csv')
```

```python
df = pd.read_csv('raw_extracted_data.csv')
df = df.drop('Unnamed: 0', axis=1)
df.columns
```

```python
# get_features.word_polarities = {}
length = len(urls)
start = 1
# length = 20
save_thres = 200
df = pd.DataFrame()
for i in range(start, length):
    print(str(i+1) + '/' + str(length))
    try:
        entry = get_features_from_article_data(scrape_article(make_valid_mashable_url(urls[i])), do_flair=False)
        entry.insert(0, "url", [urls[i]], True)
        entry.insert(len(entry.columns), 'shares', y[i])
        df = pd.concat([df, entry], ignore_index=True)
    except:
        print('something failed idk who cares move on')
    
    if i % save_thres == 0 and i != 0:
        print('Saving...')
        big = pd.read_csv('raw_extracted_data.csv')
        big = big.drop('Unnamed: 0', axis=1)
        best = pd.concat([big, df], ignore_index=True)
        best.to_csv('raw_extracted_data.csv')
        print('shape: ' + str(best.shape))
        big = None
        best = None
        df = pd.DataFrame()
```
